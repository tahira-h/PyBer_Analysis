Scatter Plot (Using Matlab Method)

```python
# Scatter plots visually depicts a negative, positive, or no-correlation relationship between two factors. 
```

```python
# Two ways to create scatter plots using the Matlab approachS

# 1. plt.plot()
# 2. plt.scatter()
```

```python
# Start with plt.plot()

# Use the same ride-sharing data
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month. 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# If we only use the plt.plot() function with our x-axis and y-axis
# it would create a line plot (Not what we are looking for. We need a scatter plot.)

# When we use the plt.plot() function to create a scatter plot. 
# add a lowercase "o" as a parameter inside the parenthesis to switch the plot from line chart to scatter plot
```

```python
plt.plot(x_axis, y_axis, 'o')
```

Create a Bubble Chart

```python
# Bubble chart (using the Matlab Method)
```

```python
# Bubble charts are used to present financial, population, and weather data

# You can add a third and fourth factor to convey more information. 
```

```python
# First two factors: 
# 1. x-axes
# 2. y-axes

# Third factor:
# 3. size

# NOTE: By changing the "dot" into a "bubble", we are adding a third factor.

# Fourth Factor: 
# 4. color

# NOTE: If there is more than one dataset using the same axes, then we change the color of each marker for each data set. 
```

Create a scatter plot 

```python
plt.scatter(x_axis, y_axis)
```

```python
# Change the size of our marker on the graph 
# Use s, the size parameter.

# Make the size of each marker equal to the data floating-point decimal values of the y-axis by using 
# s=y_axis
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# The marker size changed based on the value of the fare in USD "above"
```

```python
# Let's adjust the the size by multiplying the data in the y-axis by any value.
```

```python
# Multiply each data point in the y-axis by 3.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# Use the new y-axis list for the s parameter
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

Refactor the code

```python
# Instead of using the for loop to create the list y_axis_larger for size
# Use Python's list comprehension technique inside the plt.scatter() function

# Yes, it is approved to use list comprehension to replace many for and while loops.
```

```python
# Format for list comprehension "below"
```

```python
# new_list = [expression for item in list if conditional]
```

```python
# Use the list comprehension technique to multiply each data point in the y-axis list.
# using the plt.scatter() function
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

Pie Charts

```python
# ALlow us to depict percentages of categories as wedges of a pie.
```

```python
# Limitation?
# Can only represent one dataset or category. 
```

```python
# Pie Chart (MatLab method)
```

```python
# Use plt.pie() function to create a pie chart
```

```python
# Use the x_axis data(months) as labels, 
# and y-axis(total fare for each month) as values

plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Add percentages to the wedges, add specific colors, and explode "pop out" one or more wedges 
```

```python
# Add percentages for each month
# Explode the largest percentage, which is July, the seventh value in the x_axis.

# The 'explode' parameter will offset the indicated wedge by a fraction of the radius,
# where "0" is zero distance from the center of the pie
# and "1" is completely outside of the diameter of the pie. 
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Increase the size of the pie chart
plt.subplots(figsize=(8, 8))
```

```python
# Add the "colors" parameter in the pie() function with our other parameters
```

```python
# First assign a "colors" variable to an array with names of our 12 colors
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
       explode=explode_values,
       colors=colors,
       labels=x_axis,
       autopct='%.1f%%')

plt.show()
```
